```python
import matplotlib.pyplot as plt
import libertem.api as lt
import numpy as np
```

```python
from libertem.analysis import SumAnalysis
```

```python
ctx = lt.Context()
```

```python
path = "/home/abi/Documents/LiberTEM_data/calibrationData_circularProbe.h5"
ds_type = "HDF5"
ds_path = "4DSTEM_experiment/data/datacubes/polyAu_4DSTEM/data"
```

```python
ds = ctx.load(ds_type, path=path, ds_path=ds_path)
```

```python
sum_roi = SumAnalysis(dataset=ds, parameters={'roi': {'shape': 'rect', 'x': 32.59932513058357, 'y': 41.02833649473163, 'width': 18.881844380403464, 'height': 25.916431437995314}})
```

```python
udf = sum_roi.get_udf()
roi = sum_roi.get_roi()

sum_res = ctx.run_udf(dataset=ds, udf=udf, roi=roi, progress=True)
```

```python
plt.figure()
plt.imshow(sum_res['intensity'].raw_data)
```
