```python
import matplotlib.pyplot as plt
import libertem.api as lt
import numpy as np
```

```python
from libertem.analysis import SumAnalysis, RingMaskAnalysis
```

## Connection

***Some documentation***

```python
cluster = executor.dask.DaskJobExecutor.connect("tcp://localhost:8786")
```

```python
ctx = lt.Context(executor=cluster)
```

## Loading Dataset

```python
path = "/home/abi/Documents/LiberTEM_data/calibrationData_circularProbe.h5"
ds_type = "HDF5"
ds_path = "4DSTEM_experiment/data/datacubes/polyAu_4DSTEM/data"
```

```python
ds = ctx.load(ds_type, path=path, ds_path=ds_path)
```

## Sum Analysis

```python
sum_analysis = SumAnalysis(dataset=ds, parameters={'roi': ''})

sum_res = ctx.run(sum_analysis, progress=True)
```

```python
plt.figure()
plt.imshow(sum_res['intensity'].visualized)
```

## Ring Mask Analysis

```python
ring_analysis = RingMaskAnalysis(dataset=ds, parameters={'shape': 'ring', 'cx': 125, 'cy': 125, 'ri': 62.5, 'ro': 125})

ring_res = ctx.run(ring_analysis, progress=True)
```

```python
plt.figure()
plt.imshow(ring_res['intensity'].visualized)
```
